## Runge-Kutta 4th Order
$y^\prime=-y+t+1$

$0\leq t\leq 1$

$y(0)=1$

$w_0=y(t_0)=\alpha=1,~~~~~,~~~~~N=50,~~~~ h=\frac{b-a}{N}=\frac{1-0}{50}$

$k_1=hf(t_i,w_i)$

$k_2=hf(t_i+\frac{h}{2},w_i+\frac{k_1}{2})$

$k_3=hf(t_i+\frac{h}{2},w_i+\frac{k_2}{2})$

$k_2=hf(t_i+h,w_i+k_3)$

$w_{i+1}=w_i+\frac{1}{6}\left(k_1+2k_2+2k_3+k_4\right)$

```python
import numpy as np
a = 0
b = 1
N = 50
h = (b-a)/N
t = np.zeros(N+1)
w = np.zeros(N+1)
t[0] = 0
w[0] = 1
def f(t,y):
  d = -y+t+1
  return d
for i in range(N):
  k1 = h*f(t[i],w[i])
  k2 = h*f(t[i]+h/2,w[i]+k1/2)
  k3 = h*f(t[i]+h/2,w[i]+k2/2)
  k4 = h*f(t[i]+h,w[i]+k3)
  w[i+1] = w[i]+(1/6)*(k1+2*k2+2*k3+k4)
  t[i+1] = t[i]+h


```
