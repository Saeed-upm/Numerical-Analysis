# روش اویلر

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
def f(x,y):
  d = y
  return d
h = 0.05
N = 100
x = np.zeros(N+1)
y = np.zeros(N+1)
x[0] = 0; y[0] = 1;
for i in np.arange (N):
  y[i+1] = y[i]+h*f(x[i],y[i])
  x[i+1] = x[0] + (i+1)*h
```

```python
yexact=np.exp(x)
plt.figure(figsize = (12, 8))
plt.plot(x, yexact, label='Exact Solution')
plt.plot(x,y, 'bo--', label ='Approximation')
plt.title('Approximate and Exact Solution for Simple ODE')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='upper right')
plt.show()
```

## روش اویلر مرتبه دو

```python
import numpy as np
import matplotlib.pyplot as plt
h = 0.1
N = 10
x = np.zeros(N+1)
ye2 = np.zeros(N+1)
x[0] = 0; ye2[0] = 1;
def euler_2(x,y,h):
  d = (y-x)*(h**2/2-h)+y+h
  return d
for i in np.arange (N):
  ye2[i+1] = euler_2(x[i],ye2[i],h)
  x[i+1] = x[0] + (i+1)*h
```

## روش اویلر مرتبه چهار

```python
import numpy as np
import matplotlib.pyplot as plt
h = 0.1
N = 10
x = np.zeros(N+1)
ye4 = np.zeros(N+1)
x[0] = 0; ye4[0] = 1;
def euler_4(x,y,h):
  d = y+h*((y-x)*(-1+h/2-h**2/6+h**3/24)+1)
  return d
for i in np.arange (N):
  ye4[i+1] = euler_4(x[i],ye4[i],h)
  x[i+1] = x[0] + (i+1)*h
```

```python
plt.plot(x,ye2, 'bo--', label ='Approximation Euler-2')
plt.plot(x,ye4, 'r*--', label ='Approximation Euler-4')
plt.title('Approximate Solution for Simple ODE')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
